```python
num_rating = rating_with_df.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={
    'Book-Rating': 'num_of_rating'}, inplace= True)
```

```python
num_rating.head()
```

```python
rating_with_df.head()
```

```python
final_rating = rating_with_df.merge(num_rating , on = "Title")
```

```python
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
final_rating =final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.sample(10)
```

```python
final_rating.drop_duplicates(['user-id', 'Title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user-id', index='Title', values= 'Book-Rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
distance
```

```python
suggestion
```

```python
for i in range (len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model,open('artifact/model.pkl','wb'))
pickle.dump(books_name,open('artifact/book_name.pkl','wb'))
pickle.dump(final_rating,open('artifact/final_rating.pkl','wb'))
pickle.dump(book_pivot,open('artifact/book_pivot.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "The Sands of Time"
recommend_book(book_name)
```
